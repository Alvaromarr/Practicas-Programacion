```python
#Ejercicio 1

def processtext(char1="This is a default text with Python and amazing words.",replace=["Python","amazing"]):

    processed=char1.lower()
    processed = processed.strip()

    replacecount=0

    for word in replace:

        wordl=word.lower()

        if wordl in processed:

            processed=processed.replace(wordl,"*"*len(wordl))
            replacecount +=1

    return processed, replacecount
    



phrase, count=processtext()
print("Processed text: ",phrase)
print("Changed words: ",count)

```

```python
#Ejercicio 2

def is_prime(num):

    if not num==int(num) or num<=1:
        return False
    if num <=3 and num>1:
        return True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def prime_list(max):
    
    plist=[]

    for num in range(2,max+1):
        if isprime(num):
            plist.append(num)
    return plist
def check_palindrome(primes):
    
    palindromes = []

    for num in primes:
        if str(num) == str(num)[::-1]:
            palindromes.append(num)
    return palindromes

def categorize_prime(prime):
    
    if not is_prime(prime):
        return
    if prime<10:
        return "pequeño"
    elif prime<100:
        return "mediano"
    else:
        return "grande"
```

```python
# Ejercicio 3

def suma_total(lista):
    return sum(lista)


def suma_positivos_negativos(lista):
    positivos = sum(x for x in lista if x > 0)
    negativos = sum(x for x in lista if x < 0)
    return positivos, negativos


def max_min(lista):
    return max(lista), min(lista)


def son_unicos(lista):
    return len(lista) == len(set(lista))


def mas_positivos_o_negativos(lista):
    positivos = sum(1 for x in lista if x > 0)
    negativos = sum(1 for x in lista if x < 0)
    if positivos > negativos:
        return "Hay más positivos"
    elif negativos > positivos:
        return "Hay más negativos"
    else:
        return "Iguales"


def main():
    
    lista = input("Introduce números separados por espacios: ").split()
    lista = [int(x) for x in lista]
    
    while True:
        print("\n--- MENÚ ---")
        print("1. Sumar todos los números")
        print("2. Sumar positivos y negativos por separado")
        print("3. Encontrar máximo y mínimo")
        print("4. Comprobar si todos los números son únicos")
        print("5. Más positivos o negativos")
        print("6. Salir")
        
        opcion = input("Elige una opción: ")
        
        if opcion == "1":
            print("Suma total:", suma_total(lista))
        elif opcion == "2":
            pos, neg = suma_positivos_negativos(lista)
            print("Suma de positivos:", pos)
            print("Suma de negativos:", neg)
        elif opcion == "3":
            maximo, minimo = max_min(lista)
            print("Número máximo:", maximo)
            print("Número mínimo:", minimo)
        elif opcion == "4":
            print("Todos los números son únicos:", son_unicos(lista))
        elif opcion == "5":
            print(mas_positivos_o_negativos(lista))
        elif opcion == "6":
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida, intenta de nuevo.")


main()
```

```python
#ejercicio 4

tasks = []

def add_task(task):
    global tasks
    tasks.append(task)
    print(f"Tarea '{task}' agregada.")

def remove_task(task):
    global tasks
    if task in tasks:
        tasks.remove(task)
        print(f"Tarea '{task}' eliminada.")
    else:
        print(f"Tarea '{task}' no encontrada.")

def list_tasks():
    global tasks
    print("Lista de tareas:")
    for i, task in enumerate(tasks, 1):
        print(f"{i}. {task}")

def task_manager():
    def modify_task():
        nonlocal tasks_modified
        if tasks:
            old_task = tasks[0]
            tasks[0] = f"{tasks[0]} (modificada)"
            print(f"Tarea '{old_task}' modificada a '{tasks[0]}'.")
        else:
            print("No hay tareas para modificar.")

    tasks_modified = True

    add_task("Comprar leche")
    add_task("Llamar al doctor")
    add_task("Hacer ejercicio")
    
    list_tasks()
    
    modify_task()
    
    list_tasks()
    
    remove_task("Hacer ejercicio")
    
    list_tasks()

def main():
    task_manager()
main()
```

```python
import re
from datetime import datetime

def procesar_informe(informe):
    
    patron_fecha = re.compile(r"\b(\d{2})/(\d{2})/(\d{4})\b")

    
    nacimiento_match = re.search(r"Fecha de nacimiento:\s*(\d{2}/\d{2}/\d{4})", informe)
    if nacimiento_match:
        fecha_nac_str = nacimiento_match.group(1)
        dia, mes, anio = map(int, fecha_nac_str.split("/"))
        fecha_nac = datetime(anio, mes, dia)
        
        
        hoy = datetime.now()
        edad = hoy.year - fecha_nac.year - ((hoy.month, hoy.day) < (fecha_nac.month, fecha_nac.day))
        
        
        informe = re.sub(r"Fecha de nacimiento:\s*\d{2}/\d{2}/\d{4}", f"Edad: {edad}", informe)

    
    informe = patron_fecha.sub("FECHA", informe)

    
    informe = re.sub(r"(Nombre:\s*)(.+)", r"\1PACIENTE", informe)

    
    informe = re.sub(r"Dr\.|Dra\..*", "MEDICO", informe)

    return informe


def main():
    informe = """Informe clínico de Urgencias
Nombre: Juan Pérez López
Género: Hombre
Fecha de nacimiento: 12/05/1980
Motivo de consulta: dolor abdominal 12/05/2023
Dr. Ramírez
Tratamiento: reposo y analgésicos"""

    print("Informe original:\n")
    print(informe)
    print("\nInforme anonimizado:\n")
    print(procesar_informe(informe))



main()
```
